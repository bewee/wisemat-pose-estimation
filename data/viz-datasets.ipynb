```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from lib.constants import set_skeleton_format
from lib.viz import plot_results
```

```python
def viz(df):
    fig, axes = plt.subplots(4, 3)
    fig.set_size_inches(20, 35)
    for i in range(3):
        for j in range(3):
            index = 3*i+j
            plot_results(np.load(df.iloc[index]["pressure_image_ref"]+".npy"), [df.iloc[index]["skeleton"]], ax=axes[i][j])
    axes[3][0].hist(df['mass'])
    axes[3][0].set_title('mass [kg]')
    axes[3][1].hist(df['height'])
    axes[3][1].set_title('height [cm]')
    axes[3][2].hist(df['male'].astype(np.uint8))
    axes[3][2].set_title('sex')
    axes[3][2].set_xticks((0.05, 0.95))
    axes[3][2].set_xticklabels(('    female', 'male    '))
```

```python
set_skeleton_format('clever')
df = pd.read_pickle('bodies-at-rest/full-data.pkl')
df['pressure_image_ref'] = df['pressure_image_ref'].map(lambda x: f"bodies-at-rest/{x}")
viz(df)
df
```

```python
set_skeleton_format('slp')
df = pd.read_pickle('SLP/full-data.pkl')
df['pressure_image_ref'] = df['pressure_image_ref'].map(lambda x: f"SLP/{x}")
viz(df)
df
```
